```python
import pandas as pd
```

```python
df= pd.read_csv('../data/owid-covid-data.csv')
df.tail()
```

```python
df_new = pd.DataFrame(columns=["continent", "location", "weekly_hosp_admissions_per_million", "median_age", "aged_65_older", "aged_70_older", "life_expectancy"])
df_new= df.groupby(['continent', 'location'])['weekly_hosp_admissions_per_million', 'median_age', "aged_65_older", "aged_70_older", "life_expectancy"].max().reset_index()
df_new
```

```python
df_new=df_new.dropna()
df_new
```
