```python
import pandas as pd
import datetime as dt
```

```python
df= pd.read_csv('../data/owid-covid-data.csv')
df.tail()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)

df.tail()
```

```python
df_new= df.groupby(['continent','location', 'date'])['new_cases'].max().reset_index().fillna(0)
df_new
```

```python
df_1= df.groupby(['continent','location'])['new_cases'].max().reset_index()
df_1
```

```python
df_2 = df_1[df_1['new_cases'].isna()]
list1 = df_2['location'].tolist()
list1
```

```python
df_final=df_new[df_new.location.isin(list1) == False]
```

```python
df_final
```

```python
df_final.to_csv("../data_clean/viz7.csv")
```

```python
df_2= df.groupby(['continent', 'date'])['new_cases'].sum().reset_index()
df_2
```

```python
df_2.to_csv("../data_clean/viz7_cont.csv")
```
