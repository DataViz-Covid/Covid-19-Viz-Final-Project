```python
import pandas as pd
```

```python
df= pd.read_csv('../data/owid-covid-data.csv')
df.tail()
```

```python
df_new = pd.DataFrame(columns=["continent", "location", "total_cases", "population_density"])
df_new= df.groupby(['continent', 'location'])['total_cases', 'population_density', 'total_deaths_per_million'].max().reset_index()
df_new
```

```python
df_new=df_new.dropna()
```

```python
df_new
```

```python
df_new=df_new.replace('United States','United States of America')
df_new=df_new.replace('Democratic Republic of Congo','Democratic Republic of the Congo')
df_new=df_new.replace('Congo','Republic of Congo')
df_new=df_new.replace("Cote d'Ivoire",'Ivory Coast')
df_new=df_new.replace("Guinea-Bissau",'Guinea Bissau')
df_new=df_new.replace("Tanzania",'United Republic of Tanzania')
df_new=df_new.replace("Eswatini",'Swaziland')
df_new=df_new.replace("Serbia",'Republic of Serbia')
df_new=df_new.replace("North Macedonia",'Macedonia')
df_new=df_new.replace("Czechia",'Czech Republic')
```

```python
df_new.to_csv("../data_clean/viz1.csv")
```
