```python
import json
import pandas as pd
import math
```

```python
# Reading Countries data
with open('ne_110m_admin_0_countries.json') as file:
    for line in file:
        json_obj = json.loads(line)
        
country_objects = json_obj['objects']['countries']['geometries']
```

```python
# GDP data
df = pd.read_csv('world_gdp.csv')
```

```python
# Data manip
output_countries = []

for country in country_objects:
    country_properties = country['properties']
    country_code = country_properties['SU_A3']
    
    country_gdp = df[df['Country Code'] == country_code]
    
    gdp_2019 = None
    gdp_2021 = None
    
    default_val = -1
    
    if country_gdp.shape[0] > 0:
        gdp_2019 = int(country_gdp["2019"]) if not (math.isnan(country_gdp["2019"])) else default_val
        gdp_2021 =  int(country_gdp["2019"]) if not (math.isnan(country_gdp["2019"])) else default_val
        print(f'Country: {country_code}: 2019: {gdp_2019}; 2021: {gdp_2021}')
        
    else:
        print(f'Country: {country_code}: No data')
        
    country_properties['gdp_2019'] = gdp_2019 if gdp_2021 and not math.isnan(gdp_2019) else default_val
    country_properties['gdp_2021'] = gdp_2021 if gdp_2021 and not math.isnan(gdp_2021) else default_val
        
    country['properties'] = country_properties
    output_countries.append(country)
    
json_obj['objects']['countries']['geometries'] = output_countries
```

```python
# Output to json
jsonString = json.dumps(json_obj)
with open("gdp_map.json", "w") as file:
    file.write(jsonString)
```
