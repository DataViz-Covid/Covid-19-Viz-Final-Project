```python
import pandas as pd
import math
import json
```

```python
covid_df = pd.read_csv('owid-covid-data.csv')
covid_df
```

```python
hospital_df = pd.read_csv('current-covid-patients-hospital.csv')
hospital_df_count = hospital_df.groupby('Day')['Daily hospital occupancy'].sum()
hosp_count = int(hospital_df_count['2022-12-31'])
print(hosp_count)
```

```python
icu_df = pd.read_csv('current-covid-patients-icu.csv')
icu_df_group = icu_df.groupby('Day')['Daily ICU occupancy'].sum()
icu_count = int(icu_df_group['2022-12-31'])
print(icu_count)
```

```python
latest_country_data = covid_df.groupby('location')['date'].max()

useful_data = pd.merge(covid_df, latest_country_data, on=['location', 'date'], how='inner')

useful_data
```

```python
out = useful_data[useful_data['location'].isin(['Asia', 'Africa', 'Europe', 'North America', 'South America', 'Oceania'])][['location', 'population', 'total_deaths', 'total_cases']]
```

```python
total_cases = int(out['total_cases'].sum())
print(total_cases)
```

```python
links = []
nodes = []

i = 0

nodes.append({'node': i, 'name': 'healthy'})
health_id = i
i += 1

nodes.append({'node': i, 'name': 'contaminated'})
cont_id = i
i += 1
        
nodes.append({'node': i, 'name': 'dead'})
dead_id = i
i += 1

for index, row in out.iterrows():
    nodes.append({'node': i, 'name': row['location']})
    loc_id = i
    i += 1
    
    if not math.isnan(row['population']):
        
        if not math.isnan(row['total_cases']):
            
            links.append({
                'source': loc_id,
                'target': cont_id,
                'value': int(row['total_cases'])
            })

        if not math.isnan(row['total_deaths']):
            
            links.append({
                'source': loc_id,
                'target': dead_id,
                'value': int(row['total_deaths'])
            })

        links.append({
            'source': loc_id,
            'target': health_id,
            'value': int(row['population'] - 
                (row['total_deaths'] if not math.isnan(row['total_deaths']) else 0) - 
                (row['total_cases'] if not math.isnan(row['total_cases']) else 0))
        })
        
nodes.append({'node': i, 'name': 'At home'})
links.append({
    'source': cont_id,
    'target': i,
    'value': total_cases - icu_count - hosp_count
})
i += 1

nodes.append({'node': i, 'name': 'At Hospital'})
links.append({
    'source': cont_id,
    'target': i,
    'value': hosp_count
})
i += 1

nodes.append({'node': i, 'name': 'ICU'})
links.append({
    'source': cont_id,
    'target': i,
    'value': icu_count
})
i += 1

output = {
    'nodes': nodes,
    'links': links
}

# Output to json
jsonString = json.dumps(output)
with open("sankley.json", "w") as file:
    file.write(jsonString)
```
